```python
from pathlib import Path
import pandas as pd
import json
pd.options.display.float_format = '{:,.2f}'.format

```

```python
base_dir = Path(".")

for dir_ in base_dir.iterdir():
    print(dir_)
```

```python
exp_dir =  base_dir / "exp_by_year_tok2vec" / "nlp_project"
print(f"{exp_dir}")

assets_dir = exp_dir / "assets"
training_dir = exp_dir / "training"

train_names_metrics_agg_csv = training_dir / "train_names_metrics_agg.csv"
train_names_metrics_csv = training_dir / "train_names_metrics.csv"
test_names_metrics_agg_csv = training_dir / "test_names_metrics_agg.csv"
test_names_metrics_csv = training_dir / "test_names_metrics.csv"
metrics_json = training_dir / "metrics.json"

# for dir_ in training_dir.iterdir():
#     print(dir_)
print(f"{metrics_json}")
with open(metrics_json, "r") as f:
  data = json.load(f)

print(json.dumps(data, indent=2))
print(json.dumps(data["ents_per_type"]["group_name"], indent=2))

display(pd.read_csv(test_names_metrics_agg_csv)[['tp', 'fp', 'fn', 'precision', 'recall', 'f1', 'acc']])
display(pd.read_csv(test_names_metrics_csv).sort_values("tp", ascending=False).head(10))
display(pd.read_csv(test_names_metrics_csv).sort_values("fp", ascending=False).head(10))
display(pd.read_csv(test_names_metrics_csv).sort_values("fn", ascending=False).head(10))
```
